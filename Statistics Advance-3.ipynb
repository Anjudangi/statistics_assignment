# Q1: What is Estimation Statistics? Explain point estimate and interval estimate.

```python
# Ans.1 Estimation in Statistics:
Estimation in statistics refers to the process of inferring the value of a population parameter (like the mean, proportion, or variance) based on 
sample data. Since it is often impractical or impossible to study an entire population, statisticians rely on a sample to make inferences about the 
population as a whole.

There are two primary types of estimates:

Point Estimate
Interval Estimate
1. Point Estimate:
A point estimate is a single value or statistic that provides an estimate of a population parameter. It is derived from sample data and serves as the 
best guess for the true population value.

Example:
If you take a random sample of 100 people's heights from a large population and calculate the sample mean, that sample mean is a point estimate of the
true population mean height.
If the sample mean is 170 cm, the point estimate for the population mean is 170 cm.
Common Point Estimates:
The sample mean x
 ) is a point estimate of the population mean (Œº).
The sample proportion (ùëù^) is a point estimate of the population proportion (p).
While point estimates are easy to compute and interpret, they do not give any indication of the uncertainty or reliability of the estimate.

2. Interval Estimate:
An interval estimate provides a range of values within which the population parameter is expected to lie. This range is accompanied by a confidence 
level (usually 95% or 99%), indicating how certain we are that the true population parameter falls within that interval.

Confidence Interval (CI): The most common form of interval estimate is the confidence interval. It provides an interval estimate for a population
parameter based on sample data.
Formula for Confidence Interval for the Population Mean:
CI= xÀâ¬±Z(ns)
  where:
  ùë•Àâ= sample mean (point estimate),

Z = critical value from the standard normal distribution corresponding to the desired confidence level (e.g., 1.96 for 95% confidence),
s = sample standard deviation,
n = sample size.

Example:

Suppose a researcher calculates a sample mean height of 170 cm from 100 people and constructs a 95% confidence interval. The confidence interval might be 
(167cm,173cm)
(167cm,173cm). This means that the researcher is 95% confident that the true population mean height lies between 167 cm and 173 cm.
Why Interval Estimates are Important:
Interval estimates provide more information than point estimates because they account for the variability in the sample data and the uncertainty of
estimating the population parameter.
The wider the interval, the less precise the estimate, but the more confident you are that it contains the true population parameter.

 Summary:
Point estimation gives a single value as an estimate of a population parameter.
Interval estimation gives a range (confidence interval) that likely contains the true population parameter, providing more information about the
uncertainty of the estimate.    
```

# Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.

```python
# Ans.3 What is Hypothesis Testing?
Hypothesis testing is a statistical method used to make inferences or draw conclusions about a population based on sample data. It involves formulating 
an assumption (hypothesis) about a population parameter and then using sample data to test whether that assumption is reasonable or should be rejected.

The process of hypothesis testing is used to determine whether there is enough statistical evidence in a sample of data to support a particular belief
(hypothesis) about the population.

Steps in Hypothesis Testing:
State the null and alternative hypotheses:

Null Hypothesis (H‚ÇÄ): This is the assumption that there is no effect, no difference, or no relationship. It represents the default or status quo 
assumption.
Alternative Hypothesis (H‚ÇÅ or Ha): This is what you are trying to prove. It suggests that there is an effect, a difference, or a relationship.
Choose a significance level (Œ±):

This is the probability of rejecting the null hypothesis when it is true (Type I error). Common significance levels are 0.05 or 0.01.
Collect data and calculate a test statistic:

Based on the sample data, calculate a test statistic (e.g., t-statistic, z-statistic) that measures the difference between the observed sample results
and what is expected under the null hypothesis.
Determine the p-value or critical value:

The p-value is the probability of observing the test results under the null hypothesis. If the p-value is smaller than the significance level (Œ±),
you reject the null hypothesis.
Alternatively, you can compare the test statistic to a critical value from a probability distribution.
Make a decision:

If the p-value is less than or equal to the chosen significance level (e.g., 0.05), reject the null hypothesis.
If the p-value is greater than the significance level, fail to reject the null hypothesis.

Example:
Suppose a company claims that the average time to service a customer is 10 minutes. You collect data on 50 customers and want to test if the average 
service time is different from 10 minutes.

Null Hypothesis (H‚ÇÄ): The average service time is 10 minutes (Œº=10).
Alternative Hypothesis (H‚ÇÅ): The average service time is not 10 minutes (ùúá‚â†10).
If your p-value from the data is smaller than 0.05, you reject the null hypothesis and conclude that the average service time is significantly 
different from 10 minutes.

Summary:
Hypothesis testing is a method used to make decisions or inferences about population parameters based on sample data.
It involves formulating a null hypothesis (no effect) and an alternative hypothesis (an effect exists) and using sample data to decide which is more 
likely.
Hypothesis testing is critical in many fields, including science, medicine, business, and manufacturing, because it provides a systematic and objective
way to make decisions based on data.    


To create a hypothesis comparing the average weight of male and female college students, we need to establish a null hypothesis (H‚ÇÄ) and an alternative
hypothesis (H‚ÇÅ).
In this case, we're interested in determining if the average weight of male college students is greater than the average weight of female college
students.

Hypothesis Statements:
Null Hypothesis (H‚ÇÄ):

The null hypothesis assumes there is no difference or that the average weight of male college students is less than or equal to the average weight of female college students.
ùêª
0H: 
ùúá male ‚â§ ùúá female Œº male ‚â§Œº female
‚ÄãAlternative Hypothesis (H‚ÇÅ):

The alternative hypothesis assumes that the average weight of male college students is greater than the average weight of female college students.
1H:ùúá
male>ùúá
female Œº male >Œº female
‚ÄãWhere:
ùúá male = Population mean weight of male college students,
ùúá female= Population mean weight of female college students.
Type of Test:
Since we're testing whether the average weight of male students is greater than that of female students, this is a one-tailed hypothesis test.
Possible Test:
You can collect sample data on the weights of male and female college students and use a two-sample t-test to test the hypothesis. The t-test 
would allow you to compare the means of the two groups to determine if the difference is statistically significant.

If the p-value is smaller than your significance level (e.g., 0.05), you can reject the null hypothesis and conclude that the average weight of male 
students is likely greater than that of female students.
If the p-value is larger, you fail to reject the null hypothesis and cannot conclude that there is a significant difference.
This hypothesis test will help you determine if the assumption about male students having a higher average weight than female students holds true
based on the sample data.
```
